```python
import os
import yaml

class Config:
    def __init__(self, yaml_path: str = "data/sources.yaml"):
        with open(yaml_path, "r", encoding="utf-8") as f:
            self.sources = yaml.safe_load(f)

        # Đọc từ biến môi trường
        self.OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
        self.DEFAULT_VECTORSTORE_PATH = os.getenv("VECTORSTORE_PATH", "vectorstore/")
        self.LLM_MODEL = os.getenv("LLM_MODEL", "mistral-7b-instruct")
config = Config()
```

```python
text = """
ipcc:
  path: "data/ipcc/"
  vectorstore: "vectorstore/ipcc.index"
wmo:
  path: "data/wmo/"
  vectorstore: "vectorstore/wmo.index"
"""
with open("sources.yaml", "w") as f:
    f.write(text)
```
